The dataset contains healthcare statistics and categorical information about patients who have been diagnosed with AIDS. This dataset was initially published in 1996.

The purpose of this project is to use the existing information and history on infected patients to classify whether the new patients whose data has been recently collected are infected or not

Below are the meaning to words contained in the dataset


## Meaning of columns

> <span style="color: blue">TRT Column</span> 
*indicating the kind of treatment each patient recieved*

ZDV only (0): ZDV stands for Zidovudine, also known as azidothymidine (AZT). It was one of the first antiretroviral medications used to treat HIV. ZDV belongs to the class of drugs called nucleoside reverse transcriptase inhibitors (NRTIs).

ZDV + ddI (1): This combination includes Zidovudine (ZDV) along with another antiretroviral drug called ddI, which stands for didanosine. Didanosine is another NRTI used to treat HIV infection.

ZDV + Zal (2): This combination refers to Zidovudine (ZDV) along with another antiretroviral drug called Zalcitabine (Zal), also known as dideoxycytidine (ddC). Zalcitabine is another NRTI used in HIV treatment, though it's less commonly used now due to its side effects and the availability of newer medications.

ddI only (3): This indicates treatment with didanosine alone, without Zidovudine.

> <span style="color: blue">Age</span>

Age of patients

> <span style="color: blue">WTKG</span>

Weight per KG

> <span style="color: blue">Hemophilia (0=no, 1=yes)</span>

Refers to presence or absense of genetic disorder characterized by an inability of the blood to clot properly.

> <span style="color: blue">Homo</span>

Refers to sexual orientation of patients

> <span style="color: blue">Drugs</span>

Refers to whether an individual has a history of illicit drug use or not. eg Heroin , cocaine etc

> <span style="color: blue">Karnof</span>

The Karnofsky Performance Status Scale is a tool used to assess a patient's functional impairment and overall well-being. It's often used in healthcare settings, particularly in oncology and palliative care, to evaluate a patient's ability to perform daily activities and to guide treatment decisions.

100: Normal, no complaints, no evidence of disease.

90: Able to carry on normal activity; minor signs or symptoms of disease.

80: Normal activity with effort; some signs or symptoms of disease.

70: Cares for self; unable to carry on normal activity or to do active work.

60: Requires occasional assistance but is able to care for most personal needs.

50: Requires considerable assistance and frequent medical care.

40: Disabled; requires special care and assistance.

30: Severely disabled; hospital admission is indicated although death not imminent.

20: Very sick; hospital admission necessary; active supportive treatment necessary.

10: Moribund; fatal processes progressing rapidly.

0: Dead.

> <span style="color: blue">Oprior (Non-ZDV antiretroviral therapy pre-175 )</span>

Refers to whether an individual received antiretroviral therapy (ART) other than Zidovudine (ZDV) before the trial

> <span style="color: blue">Z30 (ZDV in the 30 days prior to 175)</span>

Refers to whether individuals received Zidovudine in the 30 days prior to trial

> <span style="color: blue">Preanti ( days pre-175 anti-retroviral therapy)</span>

Refers to the number of days before trial that an individual started antiretroviral therapy (ART) for HIV treatment.

> <span style="color: blue">Str2</span>

Refers to an individuals experience in use or exposure to antiretroviral therapy before trial

> <span style="color: blue">Strat (antiretroviral history stratification)</span>

**Antiretroviral Naive(1):** This category typically includes individuals who have never received antiretroviral therapy (ART) before. 

**> 1 but <= 52 weeks of prior antiretroviral therapy(2)**: This category includes individuals who have received antiretroviral therapy for more than 1 week but less than or equal to 52 weeks (approximately 1 year) before the current assessment or time point.

**> 52 weeks' (3)** This category includes individuals who have received antiretroviral therapy for more than 52 weeks (more than 1 year) before the current assessment or time point. 

> <span style="color: blue">Symptom</span>

**0=asymp:** This indicates that the individual is asymptomatic, meaning they are not experiencing any symptoms associated with the condition.

**1=symp:** This indicates that the individual is symptomatic, meaning they are experiencing symptoms associated with the condition.

> <span style="color: blue">Treat</span>

*What treatment patient recieved during trial*

 **treatment indicator (0=ZDV only)** Indicates that the individual used ZDV as their only medication

 **treatment indicator (1=others)** Indicates that the individual used any other treatment other than ZDV

> <span style="color: blue">Offtrt: indicator of off-trt before 96+/-5 weeks (0=no,1=yes)</span>

The analysis is focused on individuals who discontinued treatment before reaching 96 weeks (approximately 2 years) from the start of treatment, with a tolerance of plus or minus 5 weeks.

> <span style="color: blue">Cd40: CD4 at baseline</span>

Refers to a measurement of the CD4 cell count at the beginning or baseline of a study or clinical trial.

> <span style="color: blue">Cd420: CD4 at 20+/-5 weeks</span>

Refers to a measurement of the CD4 cell count taken at a specific time point, which is 20 weeks (approximately 5 months) plus or minus 5 weeks from the baseline or starting point of a study or clinical trial.

> <span style="color: blue">Cd80: CD8 at baseline</span>

 Refers to a measurement of the CD8 cell count at the beginning or baseline of a study or clinical trial."CD8" refers to a type of white blood cell called a cytotoxic T cell or CD8+ T cell. CD8 cells are an essential component of the immune system involved in recognizing and destroying infected or abnormal cells.

> <span style="color: blue">Cd820: CD8 at 20+/-5 weeksn</span>

This specifies the time frame for the CD8 measurement, which is 20 weeks from the baseline, with a tolerance of plus or minus 5 weeks. Measuring CD8 cell count at specific time intervals allows healthcare providers and researchers to monitor changes in immune function over time and assess the effectiveness of interventions or treatments.

> <span style="color: blue">Infected: is infected with AIDS (0=No, 1=Yes)</span>

Indicates whether an individual is infected with AIDS or not

```python
"""from our observation there appears to be inconsistencies in regards to the labeling of the strat and str2 column.

preanti represents the number of days a patient was exposed to treatment before trial
Strt represents history of drug use before trial and
Strat represents history of drug use an year before trial.

A preanti of 1104 should be under strat 3 because the patient was exposed to less that a year of treatment
and should have a str2 of 1 because the patient has a history of exposure to treatment.
A preanti of 0 should have a str2 of 0 and a strat of 0 but as from our observation there is a preanti of 0 with str2 and strat of 1
which is inconsistent because a patient with 0 days of exposure to treatment should have no history  of drug exposure"""


results = df[['strat' , 'str2' , 'preanti']]
results.tail(10)
```

```python
"""Correct the lableling of str2 in regards to preanti"""

def det_str2(preanti):
    if preanti == 0:
        return 0
    else:
        return 1

df['str2'] = df['preanti'].apply(det_str2)

#assert changes
preanti_str2 = df[['str2','preanti']]
preanti_str2.sample(10)
# Changes have taken effect
```

```python
"""We want to make observations of the labelling of strat in regards to preanti. A preanti of < 365 which is less than 52 weeks
should have a strat of 2. We can see the inconsitencies in the second and third variable where preanti 803 should be labeled 3 and
512 should be labeled 3 """

from pandasql import sqldf
q2 = """SELECT preanti , strat FROM df
        WHERE preanti > 365
        LIMIT 5"""
result1 = sqldf(q2)
result1
```

```python
"""A preanti of > 365 which is more than 52 weeks should havea a preanti a strat of 3.We can see inconsistencies where 224 
should be labeled 2 instead of 3 and 4 should be labeled 2 instead of 3"""

q3 = """SELECT preanti, strat FROM df
        WHERE preanti < 365
        LIMIT 5"""
result2 = sqldf(q3)
result2
```

```python
"""Make changes of the inconsistency between preanti and start"""
def det_strat(preanti):
    if preanti == 0:
        return 1
    elif preanti > 364:
        return 3
    else:
        return 2
    
df['strat']= df['preanti'].apply(det_strat)
#assert changes
strat_preanti = df[['strat', 'preanti']]
strat_preanti
# The changes have taken effect
```

```python
"""There also appears to be inconsistencies in the trt and treat column.

The trt column represents the types of treatment the patient recieved where 0 = ZDV only, 1 = ZDV + ddI, 2 = ZDV + Zal and 3 = ddI.

The treat column indicates whether the patient recieved ZDV only or a combination of ZDV and another drug where 0 represents ZDV only and 1 
represents a combination of ZDV and another drug.

Therefore if treat is 0, then trt should also be 0 because this means that the patient was only exposed to ZDV """

df[['treat','trt']]
```

```python
# correct the inconsistencies
df.loc[df['treat'] == 0, 'trt'] = 0
df.loc[df['trt'] == 0 , 'treat'] = 0 
```

```python
from pandasql import sqldf
query1= """SELECT trt, (cd420-cd40) AS 'count'
            FROM df
            WHERE infected = 1"""
result=sqldf(query1)
result.__dataframe__
result
# Calculate the average effect on T cells for each medication
average_effect_by_medication = result.groupby('trt')['count'].mean()

# Calculate the number of prescriptions for each medication
prescriptions_count_by_medication = result['trt'].value_counts()

# Normalize the effect on T cells by the number of prescriptions for each medication
normalized_effect_by_medication = average_effect_by_medication / prescriptions_count_by_medication

# Determine if the effect on T cells is negative (decline) or positive (increase) for each medication
negative_effects_by_medication = normalized_effect_by_medication[normalized_effect_by_medication < 0]
positive_effects_by_medication = normalized_effect_by_medication[normalized_effect_by_medication >= 0]

# Calculate the percentage of negative (decline) and positive (increase) effects for each medication
decline_percentage_by_medication = (negative_effects_by_medication / prescriptions_count_by_medication) * 100
increase_percentage_by_medication = (positive_effects_by_medication / prescriptions_count_by_medication) * 100

# Print the percentages for each medication
print("Percentage of negative effects (decline in T cells) by medication:")
print(decline_percentage_by_medication)
print("\nPercentage of positive effects (increase in T cells) by medication:")
print(increase_percentage_by_medication)
```
